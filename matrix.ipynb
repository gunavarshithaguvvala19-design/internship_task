```python
import numpy as np

def input_matrix():
    """Get matrix dimensions and elements from user."""
    rows = int(input("Enter number of rows: "))
    cols = int(input("Enter number of columns: "))
    print("Enter elements row-wise:")
    elements = []
    for i in range(rows * cols):
        elements.append(float(input(f"Element {i+1}: ")))
    matrix = np.array(elements).reshape(rows, cols)
    return matrix

def display_matrix(matrix, label):
    """Display matrix in structured format."""
    print(f"\n{label}:")
    print(matrix)

def main():
    print("=== Matrix Operations Tool ===")
    while True:
        print("\nOperations: 1-Add 2-Sub 3-Mul 4-Transpose 5-Det 0-Exit")
        choice = input("Choose (0-5): ")

        if choice == '0':
            print("Exiting.")
            break

        if choice in ['4', '5']:
            # Single matrix operations
            print("Enter first matrix:")
            A = input_matrix()
            if choice == '4':
                result = A.T
                display_matrix(result, "Transpose")
            elif choice == '5':
                if A.shape != (A.shape[0], A.shape[0]):
                    print("Determinant requires square matrix!")
                else:
                    det = np.linalg.det(A)
                    print(f"Determinant: {det:.2f}")
        else:
            # Two matrix operations
            print("Enter first matrix:")
            A = input_matrix()
            print("Enter second matrix:")
            B = input_matrix()

            if choice == '1':
                if A.shape == B.shape:
                    result = np.add(A, B)
                    display_matrix(result, "Addition")
                else:
                    print("Matrices must have same shape!")
            elif choice == '2':
                if A.shape == B.shape:
                    result = np.subtract(A, B)
                    display_matrix(result, "Subtraction")
                else:
                    print("Matrices must have same shape!")
            elif choice == '3':
                try:
                    result = np.dot(A, B)
                    display_matrix(result, "Multiplication")
                except ValueError as e:
                    print(f"Shape mismatch for multiplication: {e}")

if __name__ == "__main__":
    main()
67665
77657655
```
